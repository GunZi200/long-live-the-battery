```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
train_epoch_loss_data = pd.read_csv(r'tfdata\train_epoch_loss.csv')
valid_epoch_loss_data = pd.read_csv(r'tfdata\valid_epoch_loss.csv')
train_mae_remaining_cycles_data = pd.read_csv(r'tfdata\train_mae_remaining_cycle.csv')
valid_mae_remaining_cycles_data = pd.read_csv(r'tfdata\valid_mae_remaining_cycle.csv')
train_mae_current_cycles_data = pd.read_csv(r'tfdata\train_mae_current_cycle.csv')
valid_mae_current_cycles_data = pd.read_csv(r'tfdata\valid_mae_current_cycle.csv')
```

```python
train_epoch_loss = pd.DataFrame(train_epoch_loss_data,columns = ['Step','Value'])
valid_epoch_loss = pd.DataFrame(valid_epoch_loss_data,columns = ['Step','Value'])
train_mae_remaining_cycles = pd.DataFrame(train_mae_remaining_cycles_data,columns = ['Step','Value'])
valid_mae_remaining_cycles = pd.DataFrame(valid_mae_remaining_cycles_data,columns = ['Step','Value'])
train_mae_current_cycles = pd.DataFrame(train_mae_current_cycles_data,columns = ['Step','Value'])
valid_mae_current_cycles = pd.DataFrame(valid_mae_current_cycles_data,columns = ['Step','Value'])
```

```python
train_epoch_loss = train_epoch_loss.values
valid_epoch_loss = valid_epoch_loss.values
train_mae_remaining_cycles = train_mae_remaining_cycles.values
valid_mae_remaining_cycles = valid_mae_remaining_cycles.values
train_mae_current_cycles = train_mae_current_cycles.values
valid_mae_current_cycles = valid_mae_current_cycles.values
```

```python
train_epoch_loss = np.array(train_epoch_loss)
valid_epoch_loss = np.array(valid_epoch_loss)
train_mae_remaining_cycles = np.array(train_mae_remaining_cycles)
valid_mae_remaining_cycles = np.array(valid_mae_remaining_cycles)
train_mae_current_cycles = np.array(train_mae_current_cycles)
valid_mae_current_cycles = np.array(valid_mae_current_cycles)
```

```python
plt.plot(train_epoch_loss[:,0],train_epoch_loss[:,1],'r', label = 'train')
plt.plot(valid_epoch_loss[:,0],valid_epoch_loss[:,1],'b', label = 'valid',alpha=0.5)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss as a function of epochs')
plt.legend()
plt.grid()
```

```python
plt.plot(train_mae_current_cycles[:,0],train_mae_current_cycles[:,1],'r', label = 'train')
plt.plot(valid_mae_current_cycles[:,0],valid_mae_current_cycles[:,1],'b', label = 'valid',alpha=0.5)
plt.xlabel('Epochs')
plt.ylabel('mae_current_cycles')
plt.title('Mean average error of current cycles as a function of epochs')
plt.legend()
plt.grid()
```

```python
plt.plot(train_mae_remaining_cycles[:,0],train_mae_remaining_cycles[:,1],'r', label = 'train')
plt.plot(valid_mae_remaining_cycles[:,0],valid_mae_remaining_cycles[:,1],'b', label = 'valid',alpha=0.5)
plt.xlabel('Epochs')
plt.ylabel('mae_remaining_cycles')
plt.title('Mean average error of remaining cycles as a function of epochs')
plt.legend()
plt.grid()
```
